## 1. Introduction

Hence, the research aims to integrate diverse approaches to predict the hourly total demand for shared bicycles in London. It will systematically evaluate and compare the efficacy of these algorithms to ascertain their effectiveness in demand prediction.

These studies underscored the significance of forecasting sharing bike flow to optimize existing systems and served as references for cities contemplating the establishment of such systems.

In recent years, various studies have explored more complicated regression methodologies for predicting shared bicycle demand, employing a range of algorithms. For instance, Gu and Lin applied multiple linear regression and random forests (Gu, 2023), Schimohr et al. adopted XGBoost (Schimohr, 2023). While existing studies have individually demonstrated the effectiveness of various algorithms in addressing prediction problems, there is a notable lack of studies conducting extensive comparisons across a range of algorithms. Identifying the algorithm that offers the best performance and highest prediction efficiency among a diverse set of algorithms presents a valuable research opportunity.

## 5. Methodology

### 5.1 Feature Engineering
As mentioned above, methods that transform the target variable to more Gaussian distribution is needed as the transformation reduces the variance in the data, helps to reduce the effect of outliers and support further residuals analysis.

The column 'count_log' represents the transformed of 'cnt' column with log1p numpy function.

```python
bike['count_log'] = np.log1p(bike['cnt'])
plot_distribution(bike,bike.drop(['weather_code','is_holiday','is_weekend','season','hour','day_of_month','day_of_week','month','year'],axis=1).columns, kde=False)
```

The column 'is_holida' looks highly imbalanced, so grouped with "is_weekend" in a new column to represent a working day or not.

The figure in 4.4.3 shows that the daylight might be a crucial factor that is creating count differences of the data. A new column 'is_night' is created that is approximately separating the data.

```python
bike['is_non_workday'] = bike['is_holiday'] + bike['is_weekend']
bike = bike.drop(['is_holiday','is_weekend'],axis=1)
bike['is_night'] = 0
bike.loc[(bike['hour'] < 8) | (bike['hour'] > 20), 'is_night'] = 1

bike.head()
```

After splitting the timestamp column, it becomes evident that the hour and month attributes significantly influence fluctuations in bicycle usage and exhibit clear periodicity. Consequently, these attributes are encoded using sine and cosine functions to enable the model to more effectively capture temporal periodicity.

```python
def time_encode(df, column, max_value):  
    df[column + '_sin'] = np.sin(2 * np.pi * df[column] / max_value)
    df[column + '_cos'] = np.cos(2 * np.pi * df[column] / max_value)
    return df

bike_time_encode=bike.copy()
time_encode(bike_time_encode,'hour',23)
time_encode(bike_time_encode,'month',12)

bike_time_encode.head()
```

```python
df = bike_time_encode.reset_index(drop=True, inplace=False)
X = df.drop(['count_log'], axis=1)
y = df['count_log']
scaler_x = preprocessing.MinMaxScaler()
X =  pd.DataFrame(scaler_x.fit_transform(X), columns = X.columns)
```

### 5.2 Split Dataset

The dataset is separated to train and test sets, also the training set will be splitted for time series cross validation. Noticing that there is no shuffling in the data and simulates the case when the testing set is in the future. The algorithm is always trained with past data and validated with future data.

```python
def df_split(df,train_percent):
    split_index = int(train_percent * len(df))
    train = df.iloc[:split_index]
    test = df.iloc[split_index:]
    return train,test
    
X_train,X_test = df_split(X,0.7)
y_train,y_test = df_split(y,0.7)
```

### 5.3 Scoring Metric

The Root Mean Squared Logarithmic Error (RMSLE) serves as a scoring function to assess differences between predicted and actual values in contexts such as competitions and predictive modeling. This metric is particularly useful in datasets that span a broad range of values—from very large to very small—making it ideal when outliers could disproportionately skew accuracy assessments. RMSLE is favored because it minimizes the influence of outliers by focusing on percentage errors rather than absolute differences, thereby reducing sensitivity to large errors at higher value scales and increasing it at lower value scales. This quality allows RMSLE to more accurately reflect proportional inaccuracies in predictions, rather than being biased by large numerical errors.

Furthermore, RMSLE places a greater penalty on underestimations than overestimations, promoting models to avoid significant underpredictions, which can be especially critical in predictive tasks where underestimating has more severe implications than overestimating.

```python
def rmsle(y, y_pred):
    y = np.expm1(y)
    y_pred = np.expm1(y_pred)
    return np.sqrt(mean_squared_log_error( y, y_pred))


rmsle_scorer = make_scorer(rmsle, greater_is_better=False)
tscv = TimeSeriesSplit(n_splits=5)
```

```python
columns = ['weather_code','season','is_non_workday']
X_dummies = pd.get_dummies(X_train, columns=columns)
X_dummies = X_dummies.drop(['day_of_week','day_of_month','year','month'],axis=1)
```

### 5.4 Rough model test

The analysis above represents a basic test of three algorithms, specifically excluding the optimization process. The primary aim is to preliminarily assess the performance of the models, with an emphasis on observing their runtime. 

```python
def rough_test():
    names = ["LinearRegression", "Random Forrest", "XGBoost"]
    
    regressors = [
        LinearRegression(),
        RandomForestRegressor(random_state=42),
        XGBRegressor(objective ='reg:squarederror',random_state=42),  
        ]
    
    data_X = [X_dummies,X_train ,X_train ,X_dummies ,X_train ,X_train ]
    reg_columns=['algorithm','score_rmsle_mean','score_std','time']
    reg_performance = pd.DataFrame(columns=reg_columns)
    for name, regressor, X_trains in zip(names, regressors,data_X):
        time_start = time.time()
        cv_results = cross_val_score(regressor, X_trains,y_train, cv=5 ,scoring = rmsle_scorer)
        time_end = round(time.time() - time_start,3)
        mean_score = round(-cv_results.mean(),4)
        std_score = round(cv_results.std(),4)
        t= pd.DataFrame([[name,mean_score,std_score,time_end]],columns = reg_columns)
        reg_performance = pd.concat([reg_performance, t], ignore_index=True)
    return reg_performance
    
reg_performance = rough_test()
reg_performance
```

It is evident that the Ordinary Least Squares (OLS) linear regression algorithm executes rapidly. However, its prediction accuracy is significantly lower compared to the Random Forest (RF) and XGBoost algorithms. While RF and XGBoost display similar levels of accuracy, RF operates noticeably slower than XGBoost.

```python
def collect_scores(model,X_train,y_train):
    prediction_train = np.expm1((model.predict(X_train).clip(min=0)))
    y_train = np.expm1(y_train)
    rms_error = np.sqrt(mean_squared_error(y_train, prediction_train))
    mean_abs_error = mean_absolute_error(y_train, prediction_train)
    r2 = r2_score(y_train, prediction_train)
    rmsle = np.sqrt(mean_squared_log_error(y_train, prediction_train))
    return {
        "Model": model.__class__.__name__,
        "RMSE": rms_error,
        "MAE": mean_abs_error,
        "R^2": r2,
        "RMSLE": rmsle
    }

results = []
results.append(collect_scores(linear_model, X_test, y_test))
results.append(collect_scores(grid_rf.best_estimator_, X_test, y_test))
results.append(collect_scores(grid_xgb.best_estimator_, X_test, y_test))

results_df = pd.DataFrame(results)
results_df
```

### 5.6 Residual Analysis

The analysis consists of following parts:

* Normality Test: The D’Agostino’s K² test was conducted to check whether the residuals conform to a normal distribution.
* Residual Statistics: The mean, standard deviation, skewness, and kurtosis of the residuals were calculated.
* Scatter Plot: Displays the relationship between test values and residuals, checking if residuals systematically change with predicted values.
* Histogram: Used to evaluate the distribution of residuals and observe if they approximate a normal distribution.
* Q-Q Plot: Further verifies the normality of the residual distribution.

```python
def residual_plot(y_test, y_predicted,main_title):
    res =  np.expm1(y_test) - np.expm1(y_predicted)
    fig, axs = plt.subplots(1,3, figsize=(25, 5))
    stat, p = stats.normaltest(res)
    print("Performing a D’Agostino’s K^2 Test for a normal distribution of residuals.")
    print("H0 = The distribution is Gaussian.P value to reject H0 is p= 0.05 = 5%")
    print('Statistics=%.3f, p=%.5f' % (stat, p))
    
    alpha = 0.05
    if p > alpha:
        print("failed to reject H0 (H0 = The distribution is Gaussian).Can't accept any hypothesis. ")
    else:
        print('The distribution does not look Gaussian (H0 is rejected)')
    
    plt.suptitle(f'{main_title} ; mean_residual =  {res.mean():.1f}, std= {res.std():.1f},  skew =  {res.skew():.4f} , kurtosis= {res.kurtosis():.2f}'  )
    axs[0].title.set_text('Scatter Plot y_test vs residuals(true-predicted)')
    axs[0].scatter(np.expm1(y_test),res)
    axs[1].title.set_text('Histogram of residuals')
    sns.distplot(res, fit=norm, ax = axs[1]);
    axs[2].title.set_text('Q-Q plot')
    sm.qqplot(res, stats.t, distargs=(4,),line = 's', ax=axs[2])
    plt.show()

residual_plot(y_test, linear_model.predict(X_test) ,"Residual plots of OLS predictions")
residual_plot(y_test, grid_rf.best_estimator_.predict(X_test) ,"Residual plots of RandomForrest predictions")
residual_plot(y_test, grid_xgb.best_estimator_.predict(X_test) ,"Residual plots of XGBoost predictions")
```

There are some outliers that are visible in the scatter plot and histogram, but most of the quantity of residuals are near 0. The shape of residuals on the scatter plot is affected by the RMSLE metric. Underestimates have greater penalty than overestimates and the scoring is performed on percentage error - huge difference on high value predictions are not so much penalized. This metric reduces the effect of errors generated by outliers in dataset.

The residuals of all three models are not normally distributed. 

The Random forest residuals have positive skew and positive mean. Also with increasing of the predicting variable the residuals also increase, but only in positive direction. The variance of the residuals is increasing and there is a positive systematic error with increasing of the predicted variable.  

The residuals of XGBoost looks better in this terms, still there are outliers, but the mean of the residuals is closer to zero, Q-Q plot is better fitted than Random Forest.
